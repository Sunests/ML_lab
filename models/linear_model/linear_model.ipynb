# Первичная обработка данных

При первичном рассмотрении данных можно указать колонки, преобразование которых является необходимым шагом для дальнейшего анализа всего датасета.


## PassengerId

> ***PassengerId*** - разложение формата gggg_pp на 3 колонки:
1.   ***passenger_id_group*** - номер группы пассажира;
2.   ***passenger_id_number_within_group*** - номер пассажира в группе;
3.   ***passenger_id_is_within_group*** - 1, если пассажир не один в своей группе, 0 - если один.

```python
def PassengerId_processing(df: pd.DataFrame) -> pd.DataFrame:



    passId = {
        'pass_id_group': [],
        'pass_id_number': [],
        'pass_id_is_within_group': [],
    }

    for i in range(len(df['PassengerId'])):
        # Разделение значений на pass_id_group и pass_id_number
        pass_id = df['PassengerId'][i]
        pass_id = pass_id.split('_')
        pass_id = [int(x) for x in pass_id]
        passId['pass_id_group'].append(pass_id[0])
        passId['pass_id_number'].append(pass_id[1])

        # Человек находится в группе, если последующий за ним человек в списке
        # имеет pass_id_number отличный от 1.

        # Условие для последнего человека в списке.
        if i == len(df['PassengerId']) - 1:
            passId['pass_id_is_within_group'].append(1 if pass_id[1] == 2 or pass_id[1] == 3 else 0)

        else:
            # Для человека, находящегося в группе, но имеющим pass_id_number = 1
            if pass_id[1] == 1 and int(df['PassengerId'][i + 1].split('_')[1]) == 2:
                passId['pass_id_is_within_group'].append(1)

            # Для остальных людей в группе (pass_id_number = 2, 3 ...)
            else:
                passId['pass_id_is_within_group'].append(1 if pass_id[1] == 2 or pass_id[1] == 3 else 0)

    return pd.DataFrame(passId)
```
